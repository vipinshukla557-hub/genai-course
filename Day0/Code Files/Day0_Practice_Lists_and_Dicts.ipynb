### List Q4 — Loop and Count

Given this list of marks:
```python
marks = [82, 45, 91, 67, 38, 74, 55]
```

Write a loop that **counts how many students passed** (marks >= 60).
Print: `"X students passed out of Y"` 


````python
# ✍️ Your attempt here:
marks = [82, 45, 91, 67, 38, 74, 55]

````

<details>
<summary>👀 Click to see solution</summary>
</details>


````python
# ✅ Solution — List Q4

marks = [82, 45, 91, 67, 38, 74, 55]

passed = 0                        # start counter at zero
for mark in marks:                # go through each mark
    if mark >= 60:                # is this mark a pass?
        passed = passed + 1       # yes — add 1 to counter

print(f"{passed} students passed out of {len(marks)}")
# → 4 students passed out of 7
````

### List Q6 — Slice a List

Given this list:
```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
```

1. Print only the **weekdays** (first 5 items) using slicing
2. Print only the **weekend** (last 2 items) using slicing

Hint: `my_list[start:end]` gives items from `start` up to (but not including) `end`.


````python
# ✍️ Your attempt here:
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

````

<details>
<summary>👀 Click to see solution</summary>
</details>


````python
# ✅ Solution — List Q6

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

weekdays = days[0:5]     # items at index 0,1,2,3,4
weekend  = days[5:]      # items from index 5 to the end (also works: days[-2:])

print(f"Weekdays: {weekdays}")   # → ["Mon", "Tue", "Wed", "Thu", "Fri"]
print(f"Weekend:  {weekend}")    # → ["Sat", "Sun"]
````

### Dict Q4 — Loop Through a Dictionary

Given this dictionary of AI model costs (per 1K tokens):
```python
costs = {"gpt-4o": 0.005, "gpt-4o-mini": 0.00015, "claude-sonnet": 0.003}
```

Loop through it and print each model and its cost in this format:
```
gpt-4o         → $0.005
gpt-4o-mini    → $0.00015
claude-sonnet  → $0.003
```


````python
# ✍️ Your attempt here:
costs = {"gpt-4o": 0.005, "gpt-4o-mini": 0.00015, "claude-sonnet": 0.003}

````

### Dict Q5 — Nested Dictionary Access

This is how an AI message looks:
```python
message = {
    "role": "assistant",
    "content": "Python is a programming language.",
    "metadata": {
        "tokens": 8,
        "model": "gpt-4o-mini"
    }
}
```

Print:
1. The `"content"` value
2. The `"model"` value (it's inside `"metadata"`)
3. A sentence: `"The assistant used gpt-4o-mini and used 8 tokens"` 


````python
# ✍️ Your attempt here:
message = {
    "role": "assistant",
    "content": "Python is a programming language.",
    "metadata": {
        "tokens": 8,
        "model": "gpt-4o-mini"
    }
}

````

<details>
<summary>👀 Click to see solution</summary>
</details>


````python
# ✅ Solution — Dict Q5

message = {
    "role": "assistant",
    "content": "Python is a programming language.",
    "metadata": {
        "tokens": 8,
        "model": "gpt-4o-mini"
    }
}

# 1. Direct access
print(message["content"])
# → Python is a programming language.

# 2. Nested access — go one level at a time
print(message["metadata"]["model"])
# → gpt-4o-mini

# 3. Full sentence using nested values
role   = message["role"]
model  = message["metadata"]["model"]
tokens = message["metadata"]["tokens"]
print(f"The {role} used {model} and used {tokens} tokens")
# → The assistant used gpt-4o-mini and used 8 tokens
````

### Dict Q6 — List of Dictionaries (The AI Pattern!)

This is EXACTLY how you send a conversation to an AI. Given:
```python
conversation = [
    {"role": "system", "content": "You are a helpful tutor."},
    {"role": "user", "content": "What is Python?"},
    {"role": "assistant", "content": "Python is a programming language."},
    {"role": "user", "content": "Why is it used in AI?"}
]
```

1. Print how many messages are in the conversation
2. Print the **first** message's content
3. Print the **last** message's content
4. Loop through and print each message as: `[ROLE]: content`


````python
# ✍️ Your attempt here:
conversation = [
    {"role": "system", "content": "You are a helpful tutor."},
    {"role": "user", "content": "What is Python?"},
    {"role": "assistant", "content": "Python is a programming language."},
    {"role": "user", "content": "Why is it used in AI?"}
]

````

<details>
<summary>👀 Click to see solution</summary>
</details>


````python
# ✅ Solution — Dict Q6

conversation = [
    {"role": "system", "content": "You are a helpful tutor."},
    {"role": "user", "content": "What is Python?"},
    {"role": "assistant", "content": "Python is a programming language."},
    {"role": "user", "content": "Why is it used in AI?"}
]

# 1. Count messages
print(f"Messages: {len(conversation)}")
# → Messages: 4

# 2. First message's content (index 0, then key "content")
print(f"First: {conversation[0]['content']}")
# → First: You are a helpful tutor.

# 3. Last message's content (index -1)
print(f"Last: {conversation[-1]['content']}")
# → Last: Why is it used in AI?

# 4. Loop through all messages
print()
for msg in conversation:
    role    = msg["role"].upper()        # make role uppercase
    content = msg["content"]
    print(f"[{role}]: {content}")
````
