## 🟢 Part 6 — Making Decisions: if / elif / else

Programs need to choose what to do. "If this is true, do this — otherwise, do that."

```python
if condition:
    do this thing
else:
    do this instead
```

⚠️ **Indentation is required.** The code inside an `if` block must be indented (4 spaces or Tab).
Python uses indentation to know what's "inside" the block.


````python
# Simple if / else
weather = "sunny"   # 🎯 try changing this to "rainy" and run again

if weather == "sunny":
    print("Wear sunglasses")
else:
    print("Take an umbrella")

print("Have a good day!")   # this always runs — it's outside the if/else
````

⚠️ `==` (double equals) means *compare*. `=` (single equals) means *store*. They look similar but do completely different things.


````python
# elif = "else if" — check more than two options
# Python checks top to bottom, stops at the first True condition

score = 72   # 🎯 Try changing this: 95, 72, 55, 30

if score >= 90:
    print(f"{score} — Excellent!")
elif score >= 75:
    print(f"{score} — Good work!")
elif score >= 60:
    print(f"{score} — Passing, keep going")
else:
    print(f"{score} — Needs more practice")
````

## 🟢 Part 7 — Loops: Doing Things Repeatedly

What if you want to print 100 names? You don't write 100 `print()` lines.
You use a **loop**.

A `for` loop says: *"for each item in this collection, do this once."*


````python
# Loop through a list
cities = ["Mumbai", "Delhi", "Bangalore", "Hyderabad"]

for city in cities:        # 'city' takes each value one by one
    print(f"📍 {city}")    # this line runs 4 times

# Loop 1: city = "Mumbai"
# Loop 2: city = "Delhi"
# Loop 3: city = "Bangalore"
# Loop 4: city = "Hyderabad" 
````

````python
# range() — repeat a specific number of times
# range(5) gives numbers: 0, 1, 2, 3, 4

for i in range(5):
    print(f"Run number {i}")
````

````python
# Loops + if/else together — very powerful combination
scores = [88, 45, 92, 61, 73]

for score in scores:
    if score >= 70:
        print(f"  {score} → Pass ✅")
    else:
        print(f"  {score} → Not yet ❌")
````

### Looping Through a Dictionary

````python
# .items() gives you key AND value together
model_costs = {
    "gpt-4o":      0.005,
    "gpt-4o-mini": 0.00015,
    "claude-sonnet": 0.003
}

for model_name, cost in model_costs.items():
    print(f"{model_name} costs ${cost} per 1K tokens")
````

### Stopping a Loop Early: `break`

Sometimes you want to stop looping once you find what you need.


````python
# break — exit the loop immediately
numbers = [10, 25, 30, 42, 55, 60]

for num in numbers:
    print(f"Checking {num}...")
    if num > 40:
        print(f"Found it! {num} is greater than 40")
        break   # stop the loop right here — don't check the rest
````

### Functions with Parameters (Inputs)

Make a function flexible by letting you pass information to it.


````python
# A function that takes one input (called a "parameter")
def greet(name):
    print(f"Hello, {name}! Great to have you here.")

# Call it with different names — the function adapts
greet("Aarav")
greet("Sneha")
greet("Mohammed")
````

### Default Parameters

You can give a parameter a **default value** — used when the caller doesn't provide one.


````python
# 'language' has a default value of "English"
# If you don't pass it, Python uses "English" automatically
def greet_in(name, language="English"):
    if language == "Hindi":
        print(f"Namaste, {name}!")
    else:
        print(f"Hello, {name}!")

greet_in("Priya")              # uses default: English
greet_in("Rahul", "Hindi")     # we provided Hindi, so it uses Hindi
````

### Functions that Return a Value

A function can also **give back** a result using `return`.


````python
# A function that calculates and returns a value
def estimate_tokens(text):
    # Rough rule: 1 token ≈ 4 characters in English
    token_count = len(text) // 4     # // means divide and round down
    return token_count               # send the result back

# The returned value can be stored in a variable
my_prompt = "What is machine learning?"
tokens = estimate_tokens(my_prompt)
print(f"Prompt: '{my_prompt}'")
print(f"Estimated tokens: {tokens}")
````

### Functions Calling Other Functions

Real programs are built by combining small functions — each does one thing well.


````python
# Function 1: count words
def count_words(text):
    return len(text.split())   # .split() breaks text into a list of words

# Function 2: uses Function 1 inside it
def analyse_prompt(prompt):
    words = count_words(prompt)           # calling our other function
    tokens = estimate_tokens(prompt)      # calling the function from above
    print(f"Prompt: '{prompt}'")
    print(f"  Words:  {words}")
    print(f"  Tokens: {tokens}")

analyse_prompt("Explain machine learning in simple words")
print("---")
analyse_prompt("Hi")
````

## 🟢 Part 10 — Classes: Understanding Dot Notation

You've already seen things like `math.sqrt()`, `random.choice()`, `favourites.append()`.

That **dot** (`.`) is everywhere in Python. What does it mean?

A **class** is a blueprint for creating **objects**. An object has:
- **Data** (called attributes) — things it knows
- **Actions** (called methods) — things it can do

You access both using the **dot**: `object.attribute` or `object.method()`

You don't need to write many classes — but you need to **read** them, because all AI libraries use them.


````python
# Here's a simple class — a blueprint for an AI model
class AIModel:
    def __init__(self, name, provider):
        # __init__ runs automatically when you create an object
        # 'self' means "this particular object"
        self.name = name            # store name on the object
        self.provider = provider    # store provider on the object

    def describe(self):
        # A method — an action the object can perform
        print(f"{self.name} by {self.provider}")

# Create two objects from the same blueprint
model1 = AIModel("gpt-4o-mini", "OpenAI")       # create one object
model2 = AIModel("claude-3-haiku", "Anthropic")  # create another

# Use the dot to access data and actions
print(model1.name)        # access data (attribute)
print(model2.provider)    # access data (attribute)
model1.describe()         # perform action (method)
model2.describe()         # perform action (method)
````

### Why This Matters: The OpenAI Pattern

In Week 1, you'll write code that looks like this:

```python
from openai import OpenAI          # import the class
client = OpenAI(api_key="...")     # create an object from the class
response = client.chat.completions.create(...)  # call a method using dot
```

Now you know what every part means:
- `OpenAI` is a **class** (blueprint)
- `client` is an **object** (created from the blueprint)
- `.chat.completions.create()` is a **method** (an action using the dot)


````python
# You already use objects without knowing it!
# A string IS an object with methods:

text = "hello world"
print(text.upper())         # .upper() is a method → "HELLO WORLD"
print(text.split())         # .split() is a method → ["hello", "world"]
print(text.replace("world", "AI"))  # .replace() is a method

# A list IS an object too:
items = [3, 1, 2]
items.sort()                # .sort() is a method
print(items)                # → [1, 2, 3]
````

> **Key takeaway:** Whenever you see a **dot** in Python, it means "reach inside this object and access something." That's it. You'll see this hundreds of times.

## 🟢 Part 11 — When Things Go Wrong: try / except

AI APIs fail. Networks time out. Files go missing. That's normal.

**Error handling** lets your program keep running instead of crashing.

```python
try:
    # attempt this code
except:
    # if it fails, do this instead
```

Think of it like a safety net — if the trapeze artist falls, the net catches them.


````python
# Without try/except — this CRASHES your program
# (Uncomment the next line to see the error, then re-comment it)

# print(10 / 0)   # ZeroDivisionError — Python can't divide by zero
````

````python
# With try/except — the error is caught, program continues

try:
    result = 10 / 0                    # this will fail
except:
    print("Oops! Something went wrong.")

print("But the program keeps running!")  # this line still executes
````

````python
# You can catch SPECIFIC types of errors
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:          # only catches divide-by-zero
        print("❌ Cannot divide by zero!")
        return None                     # return None instead of crashing

print(safe_divide(10, 2))    # works fine → 5.0
print(safe_divide(10, 0))    # caught gracefully → None
````

````python
# The pattern you'll use EVERY time you call an AI API:

def call_ai(prompt):
    try:
        # In real code, this is where the API call happens
        # response = client.chat.completions.create(...)
        print(f"Sending: '{prompt}'")
        print("✅ Got a response!")
        return "This is the AI's answer"
    except Exception as e:
        # 'e' contains the error message — useful for debugging
        print(f"❌ API call failed: {e}")
        return None

# Test it
answer = call_ai("What is Python?")
print(f"Answer: {answer}")
````

> **Why try/except matters:** Without it, one failed API call crashes your entire application. With it, your app handles failures gracefully. This is the difference between a demo and a real product.

## 🟢 Part 12 — Decorators: Adding Superpowers to Functions

This is the most advanced topic today — but we'll build up to it step by step.

**Big idea:** A decorator is a way to **add extra behaviour** to a function without changing the function itself.

Think of it like adding a phone case — the phone still works the same, but now it has extra protection.

### Step 1: Functions can receive other functions


````python
# In Python, you can pass a function as an input to another function

def say_hello():
    print("Hello!")

def say_goodbye():
    print("Goodbye!")

def run_twice(any_function):        # takes a function as input
    any_function()                   # call it once
    any_function()                   # call it again

run_twice(say_hello)     # pass the function itself (no parentheses!)
print("---")
run_twice(say_goodbye)
````

### Step 2: A wrapper — the core idea

A wrapper is a function that runs some extra code **before** and **after** another function.


````python
# A wrapper adds behaviour around an existing function
def add_logging(original_function):
    def wrapper():
        print(f"▶️  Starting: {original_function.__name__}")  # extra: log before
        original_function()                                     # run the real function
        print(f"✅ Finished: {original_function.__name__}")    # extra: log after
    return wrapper    # return the enhanced version

# Original function — simple
def process_data():
    print("   Processing data...")

# Wrap it — now it has logging added
enhanced = add_logging(process_data)
enhanced()   # calls the wrapper, which calls process_data inside
````

### Step 3: The `@` shortcut

Writing `enhanced = add_logging(process_data)` every time is tedious. Python gives us `@` as a shortcut — it does the exact same thing.


````python
import time   # for measuring time

# A decorator that measures how long a function takes
def measure_time(func):
    def wrapper():
        start = time.time()           # record start time
        func()                         # run the actual function
        elapsed = time.time() - start  # calculate how long it took
        print(f"⏱️  Took {elapsed:.2f} seconds")
    return wrapper

# The @ symbol is just a shortcut for: count_to_million = measure_time(count_to_million)
@measure_time
def count_to_million():
    total = 0
    for i in range(1_000_000):
        total += i
    print(f"   Sum: {total}")

count_to_million()   # automatically measures time!
````

### The Pattern You'll Use: `@retry`

In real AI work, you'll use decorators like `@retry` to automatically retry failed API calls. Here's a simplified version:


````python
# A simple retry decorator — you'll use something like this from Week 2
import random

def retry(func):
    def wrapper():
        for attempt in range(1, 4):         # try up to 3 times
            try:
                return func()                # try calling the function
            except Exception as e:
                print(f"  Attempt {attempt} failed: {e}")
                if attempt == 3:
                    print("  ❌ All attempts failed")
                    return None
    return wrapper

@retry
def call_unreliable_api():
    # Simulate an API that fails sometimes
    if random.random() < 0.7:               # 70% chance of failure
        raise ConnectionError("Network timeout")
    return "Success! Here's your AI response."

# Run it — the decorator automatically retries on failure
result = call_unreliable_api()
if result:
    print(f"✅ {result}")
````

> **Why decorators matter:**
> - FastAPI uses `@app.get("/endpoint")` to create web APIs
> - LangChain uses `@chain` to compose AI pipelines
> - You'll use `@retry` to handle flaky API calls
>
> If you understand the `@` symbol, you understand how modern Python AI frameworks work.
